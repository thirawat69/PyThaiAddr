### trigram dictionary

### bigram dictionary

```python
from pythaiaddr import Corpus_gram

# ['word1word2, word1word2, ..., word1word2 ]
bi = Corpus_gram("bi")
print(bi.words)
```

### unigram dictionary

```python
from pythaiaddr import Corpus_gram

# ['word', 'word', ..., 'word']
uni = Corpus_gram("uni")
print(uni.words)
```

### Delete dictionary

```python
from pythaiaddr import Preprocess, Corpus_gram

unigram = Corpus_gram(gram="uni")
delete_dict = Preprocess(unigram)

delete_dict.dict_corrotion
```

## **correct address**

### warp
- ediz
- norvig
- symspell

```python
from pythaiaddr import correct_addr
from pythaiaddr import word_tokenize as newmm_tokenize


addr = "ที่อยูล่ 31/631 หมูี่ที่ 16 แขวงพระบรมถหาราชวัง เขตพระนซคร จ.กรุงเทพมหานค 1113"
token = newmm_tokenize(addr)
print(token)

print(correct_addr(addr, engine='ediz'))
print(correct_addr(addr, engine='norvig'))
print(correct_addr(addr, engine='symspell'))
```
